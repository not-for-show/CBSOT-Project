# **Search [return]**

```python
def search_paper_return(query, k=5):
  # Generate query embedding
  query_embedding = model.encode([query])

  # Normalize query embedding
  faiss.normalize_L2(query_embedding)

  # Search in FAISS index
  D, I = index.search(query_embedding, k)

  return D, I
```

```python
D , I = search_paper_return("deep learning for medical image analysis")
print(D)
print(I)
```

# **Search [display - zip()]**

```python
def search_paper_display(query, k = 5):
  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)
  D, I = index.search(query_embedding, k)

  for rank, (score, idx) in enumerate(zip(D[0] , I[0]), start = 1):
      print()
      print("=" * 100)
      print()
      print(f"Result Rank        : #{rank}")

      print(f"Paper Index        : {idx}")

      print(f"Similarity Score   : {score:.4f}")

      print(f"\nTitle: {df.iloc[idx]['title']}")

      print(f"\nAbstract: {df.iloc[idx]['abstract'][:500]}")
```

```python
search_paper_display("deep learning for medical image analysis")
```

# **Search + Summarize**

```python
def search_and_summarize(query, k=5):

  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)

  D, I = index.search(query_embedding, k)

  for rank, (score, idx) in enumerate(zip(D[0], I[0]) , start = 1):
    print()
    print("=" * 100)
    print()

    print(f"Result Rank        : #{rank}")
    print(f"Paper Index        : {idx}")
    print(f"Similarity Score   : {score:.4f}")

    print(f"\nTitle:")
    print(df.iloc[idx]["title"])

    print(f"\nAbstract Preview:")
    print(df.iloc[idx]["abstract"][:500])

    # Generate summary
    summary = summarizer(df.iloc[idx]["abstract"] , max_length = 120, min_length = 40, do_sample = False)

    print("\nSummary:")
    print(summary[0]["summary_text"])
    print("\n")
```

```python
search_and_summarize("Deep learning in medical imaging" , k = 5)
```

# **Search + Summarize + Keyword**

```python
def search_summarize_keywords(query, k=5):
  query_embedding = model.encode([query])
  faiss.normalize_L2(query_embedding)
  D, I = index.search(query_embedding, k)

  for rank, (score, idx) in enumerate(zip(D[0], I[0]), start = 1):
    print()
    print("=" * 100)
    print()

    print(f"Result Rank        : #{rank}")
    print(f"Paper Index        : {idx}")
    print(f"Similarity Score   : {score:.4f}")

    print("\nTitle:")
    print(df.iloc[idx]["title"])

    print("\nAbstract Preview:")
    print(df.iloc[idx]["abstract"][:500])

    print()

    # Generate summary
    summary = summarizer( df.iloc[idx]["abstract"] , max_length = 120, min_length = 40, do_sample = False)
    print("Summary:")
    print(summary[0]["summary_text"])
    print()

    # Extract keywords from the abstract
    keywords = kw_model.extract_keywords( df.iloc[idx]["abstract"] , keyphrase_ngram_range = (1, 3), stop_words = "english")

    print("Keywords:")

    for keyword, score in keywords:
      print(keyword)
    print("\n")
```

```python
search_summarize_keywords("natural language processing using transformers", k = 5)
```
